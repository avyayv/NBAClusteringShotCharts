```python
from bs4 import BeautifulSoup
import requests
```

```python
contract_data = "https://www.basketball-reference.com/contracts/players.html"
html_data = requests.get(contract_data)
soup = BeautifulSoup(html_data.text)
```

```python
player_to_salaries = {}
for tr in (soup.find_all('tr')):
    if len(tr.find_all('td')) > 2:
        tds = tr.find_all('td')
        name = tds[0].text
        year_by_year = []
        for td in tds:
            if len(td) > 0 and "$" in td.text:
                year_by_year.append(int(td.text.replace(',', '').replace('$', '')))
        player_to_salaries[name] = year_by_year
for name in player_to_salaries:
    print(name, sum(player_to_salaries[name][0:-1])-player_to_salaries[name][-1])
```
